### Predict whether the patient has diabetes using the Pima Indian Diabetes dataset.

The features are:
1. Number of times pregnant
2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
3. Diastolic blood pressure (mm Hg)
4. Triceps skin fold thickness (mm)
5. 2-Hour serum insulin (mu U/ml)
6. Body mass index (weight in kg/(height in m)^2)
7. Diabetes pedigree function
8. Age (years)

Class label: 0 or 1 for negative for Diabetes or positive

```
import numpy as np
import tensorflow as tf
import math
import h5py
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.python.framework import ops
from keras import layers
from keras.layers import Input, Dense, Activation, BatchNormalization
from keras.models import Model
from keras.models import Sequential
```

The datafile is imported to Google Drive. Load the file from there.

```
data_file = '/content/drive/My Drive/MLData/diabetes_data.csv'

dataset = np.genfromtxt(data_file, delimiter=',')

class_labels = dataset[:, -1]
raw_data = dataset[:, :-1]

print("Size of labels : " + str(class_labels.shape))
print("Size of data : " + str(raw_data.shape))

X = raw_data
Y = class_labels
```

Define a model. We have 8 features in the input data, that is specified in the input layer. We use ReLU activation for all the hidden layers. Since it is binary classification problem, output layer has sigmoid activation.

```
def prediction_model():
  model = Sequential()
  model.add(Dense(20, input_dim=8, activation='relu'))
  model.add(Dense(15, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  
  return model
```

Use Adam Optimizer with binary cross-entropy loss.

```
diabetes_model = prediction_model()

diabetes_model.compile(loss='binary_crossentropy', optimizer="Adam", metrics=['accuracy'])
```

```
diabetes_model.fit(x = raw_data , y = Y, epochs = 500, batch_size = 64)
```

Show the summary of the model and evaluation of the model

```
diabetes_model.summary()
scores = diabetes_model.evaluate(X, Y)
print("\n%s: %.2f%%" % (diabetes_model.metrics_names[1], scores[1]*100))
```
